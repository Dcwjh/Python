# 读取数据到列表中

```python
from collections import Counter
import numpy as np

following = []
follower = []
answer =[]
voteup = []
thanked = []
with open('tongji.csv') as f:
    for line in f:
        # print(line)
        following.append(line.split(',')[0])
        follower.append(line.split(',')[1])
        answer.append(line.split(',')[2])
        voteup.append(line.split(',')[3])
        thanked.append(line.split(',')[4])     
following = following[1:]
follower = follower[1:]
following = np.sort(following)
following = following.fillna(value=0)
        
```

```python
c = Counter()
for item in following:
    c[item] += 1
c
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

data = pd.read_csv('C:\\Users\\Mr.Wang\\zhihudata\\tongji.csv')

data = data.fillna(value=0)
#print(data.describe())
data = data.sort_values(["following_count"], ascending=True)
data.to_csv("following_tongji.csv")
following_count = pd.to_numeric(data.following_count)
cfollowing = Counter()
# print(following_count)
for item in following_count:
    cfollowing[item] += 1


df = pd.DataFrame.from_dict(cfollowing, orient='index').reset_index()
df.columns = ['following_count', 'count']
print(df.head(20))

```
